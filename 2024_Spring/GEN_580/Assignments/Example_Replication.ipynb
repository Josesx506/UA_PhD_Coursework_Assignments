```python
from fracture_formulas import sig_theta1,sig_theta2
import numpy as np
import matplotlib.pyplot as plt
```

```python
crack_angle = np.arange(-90,90.2,0.2)
a = 10
beta = 45
stress = 5
k1 = stress * np.sqrt(np.pi * a) * np.cos(np.radians(beta))**2
k2 = stress * np.sqrt(np.pi * a) * np.sin(np.radians(beta)) * np.cos(np.radians(beta))
m1 = sig_theta1(k1,1,crack_angle)
m2 = sig_theta2(k2,1,crack_angle)

plt.plot(crack_angle,m1)
plt.plot(crack_angle,m2)
plt.show()
```

```python
friction_angle = 35
crack_angle = np.arange(0,90.2,0.2)
k2 = np.sin(np.radians(crack_angle)) * np.cos(np.radians(crack_angle))
k2 = k2 - (np.tan(friction_angle) * np.sin(np.radians(crack_angle))**2)

plt.plot(crack_angle,k2)
plt.show()
```
